```python
from transformers import MarianMTModel, MarianTokenizer

# Use the English to Nepali translation model
model_name = "Helsinki-NLP/opus-mt-en-hi"
```

```python
# Load the tokenizer and model
tokenizer = MarianTokenizer.from_pretrained(model_name)
model = MarianMTModel.from_pretrained(model_name)
```

```python
def translate_to_nepali(text):
    # Tokenize input
    inputs = tokenizer(text, return_tensors="pt", padding=True)

    # Generate translation
    translated = model.generate(**inputs)

    # Decode output
    hindi_text = tokenizer.decode(translated[0], skip_special_tokens=True)
    return hindi_text
```

```python
# Example usage
english_text = "Humans change with time."
hindi_translation = translate_to_nepali(english_text)
```

```python
print("English:", english_text)
print("hindi:", hindi_translation)
```
